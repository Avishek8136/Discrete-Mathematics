a.Generate and print the following four sets from a set of integers 1 to 100.

1. A = set of integers from 1 to 100 that are prime

```python
def prime(n):
  for i in range(2,n):
    if(n%i==0):
      return 0
  return 1

for i in range(2,101):
  if(prime(i)==1):
    print(i)
```

2. B = set of integers from 1 to 100 that are even

```python
def even(n):
  if(n%2!=0):
    return 1
  return 0

for i in range(1,101):
  if(even(i)==0):
    print(i)
```

3. C = set of integers from 1 to 100 that are multiples of 5

```python
def multiple(n):
  if(n%5==0):
    return 1
  return 0

for i in range(1,101):
  if(multiple(i)==1):
    print(i)
```

**b.** Write a program to generate all possible passwords using letters from the set {a, b, c, A,
B, C} and digits from the set {5,6}. The password is case sensitive, length between 5 and
7 and starts with a letter
