Question 1: Write a program to find the minimum of a set of numbers using the following pseudocode.

Lab 3.jpg

```python
def minArray(A,n):
  min=A[1]
  for c in range(n):
    if(min>A[c]):
        min=A[c]
  return min


A = []

num = int(input('How many numbers: '))

for n in range(num):
    numbers = int(input('Enter number '))
    A.append(numbers)

print("Minimum element in the list is :", minArray(A,num))

```

Question 2: Write a program to find the sum of a set of numbers.

Lab 3.jpg



```python
def SUM(A,n):
  sum=0
  for c in range(n):
    sum = sum + A[c]
  return sum


A = []

num = int(input('How many numbers: '))

for n in range(num):
    numbers = int(input('Enter number '))
    A.append(numbers)

print("Sum of all elements in the list is :", SUM(A,num))
```

Question 3: Write a program to find the sum of positive numbers in a set of numbers.
Lab 3.jpg

```python
def SUM(A,n):
  sum=0
  for c in range(n):
    if (A[c] > 0):
      sum = sum + A[c]
  return sum


A = []

num = int(input('How many numbers: '))
d= num
print("The program will only add positive numbers")
for n in range(num):
    numbers = int(input('Enter number '))
    A.append(numbers)


print("Sum of all positive elements in the list is :", SUM(A,d))
```

Question 5: Write a program to compute the factorial of a number.



```python
def factorial(a):
  n=1
  c=1
  while(n<=a):
    c=c*n
    n=n+1
  return c


print("Enter a value")
a=int(input())
print("The factorial is ", factorial(a))
```
